The num column was dropped because it represents the original diagnosis severity and is directly related to target_binary, which could lead to data leakage during model training.

```python
df.drop('num',axis=1,inplace=True)
```

defining the dependent and independent features in x and y variables

```python
x=df.iloc[:,: -1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

well lets do the train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.32,random_state=42,stratify=y)
```

well lets do the scalar or standardization on the dataset

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
regression=LogisticRegression()
regression.fit(x_train_scaled,y_train)
y_pred=regression.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
